STANDARDIZATION

Standardization is a process by which we will make our data "Unitless". Standardization doesn't change the importance of any value, it just brings all the values to small scale, so that all the columns can be equally important for our Model. It plays a major role in increasing accuracy of our model.

```python
from sklearn.preprocessing import StandardScaler
```

```python
df
```

```python
x = df[['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']]
x
```

```python
C = x.columns
C
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
x = pd.DataFrame(x, columns = C)
x
```

```python
x.describe()
```

As we can see, after applying Standardization, all the values of our column have came to small scale. Also mean of every column have been 0 and Standard Deviation of every column have been 1. 

```python
df['Volume']  = x['Volume']
df['Avg Price']  = x['Avg Price']
df['Total Sales Value']  = x['Total Sales Value']
df['Discount Rate (%)']  = x['Discount Rate (%)']
df['Discount Amount']  = x['Discount Amount']
df['Net Sales Value']  = x['Net Sales Value']
```

```python
df
```

ENCODING

-  We generally perform encoding on the categorical columns only and it is a must step in EDA as our model doesn't understand any text value included in our data, it just understands numbers. By encoding we are converting all the unique values in a categorical column as a unique column having binary values (0 and 1), which can be understood by our model. 

```python
df = pd.get_dummies(data = df, columns = ['Date', 'Day', 'SKU', 'BU', 'Brand', 'Model'])
```

For 'City' column, we have only one unique value. For columns having same values all across, it is not at all useful to learn our model, as we are giving same information for 'City' for all rows, it won't make any sense to the model and so 'City' column is not important. Therefore, we will delete the city Column. 

```python
df
```

```python
df.drop(columns = ['City'], inplace = True)
```
