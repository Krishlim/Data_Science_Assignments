```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules, apriori
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_excel(r"F:\Krishna\Excel R\Data Scientist\Assignments\Association Rules\Online retail.xlsx")
df1
```

```python
df1['Products'] = df1.iloc[:,0]
```

```python
df1
```

```python
df = df1[['Products']]
```

```python
df
```

```python
df['Products'].unique()
```

```python
def  text_split(txt):
    return txt.split(',')
```

```python
data = df['Products'].apply(text_split) 
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
te.columns_     #shows unique values encoded by transaction encoder
```

```python
data = pd.DataFrame(encoded_df, columns = te.columns_)
data
```

```python
data.replace([True, False], [1,0], inplace = True)
data  
```

```python
data.shape
```

```python
scores = apriori(data, min_support = 0.002, use_colnames = True) 
scores        #bcz we have 120 unique products with more transactions, have set low value for min support (0.002)
```

```python
rules = association_rules(scores)
rules
```

We have had only 2 rules as above with almost 95% and 84% confidence and high lift ratio, showing both of above rules are Good Associations.

Let's read both of the above rules:
1. Customers buying mushroom cream sauce and pasta will also buy escalope with 0.2% support and 95% confidence.
2. Customers buying olive oil, tomatoes and frozen vegetables will also buy spaghetti with 0.2% support and 84% confidence.

The algorithm haven't been able to give us more rules, which shows that The sell of the store (Online Retail Store) do not have many associations or repeating pattern of similar things bought together even after having 7500 rows of data. 

Interview Questions

1.  What is lift and why is it important in Association rules?

-  In association rule mining, lift is a metric that measures the strength of the association between two items in a dataset. It's calculated by dividing the confidence of the association by the support of the second item.
-  Lift is important because it helps identify strong relationships between items, which can be useful for further investigation. A lift value greater than 1 indicates that the association is stronger than expected, while a value less than 1 indicates that the association is weaker than expected. 





2.	What is support and Confidence. How do you calculate them?

-  Support indicates how frequently an item appears in the data. Confidence indicates the number of times the if-then statement is found to be true.
-  Support(A) = Frequency(A) / # Total records
-  Confidence(Aâ†’B) = Probability(A & B) / Support(A)
