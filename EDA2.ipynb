We can see we have so many columns with Text values , so first of all ,  we will encode them. For independent columns having less than 5 unique values, we will use one hot encoding. For independent columns having greater than 5 unique values, we will use Ordinal Encoding and for target column we will use label Encoding.

*  OHE (One Hot Encoding) : Pros : easy to use , used for independent columns
                         Cons : For columns having more unique values, it creates so many new columns
*  Label Encoding   : Pros : doesnt matter how many unique values are there, it encode them in a single column
                   Cons : used only for target columns 
*  Ordinal Encoding    : Pros : easy to use, doesnt create so many columns, encode them in a single column only
                      Cons  : used only for independent columns 

```python
from sklearn.preprocessing import OrdinalEncoder
```

Now, let's detect Outliers and remove them to get better accuracy. We will use Random Forest method for the same.
